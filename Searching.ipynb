```python
# Binary search
# Sorted array
# o(1)
# o(log n)

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        elif arr[mid] < target:
            left = mid + 1

        else:
            right = mid -1

    return -1

arr = [1,2,3,4,5,6,7,8,9]
print(binary_search(arr,4))
```

```python
# Jump Search

import math

def jump_search(arr, target):
    n =  len(arr)
    step = int(math.sqrt(n))
    prev = 0


    while arr[min(step,n)-1] <  target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return  -1

    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i

    return -1

arr = [1,2,3,4,5,6,7,8,9]
print(jump_search(arr,7))
```
